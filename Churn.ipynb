# Random forests with bootstrapping

### Get the independent and dependent variables

```python
#independent variables
x=churn_copy.iloc[:,[3,6,7,9,10,11]].values
#labels
y=churn_copy.iloc[:,[13]].values
```

### Split the dataset

```python
#split the dataset
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=10,random_state=1)
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
X_train.shape
```

### Do feature scaling

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train) 
X_test = scaler.transform(X_test)
```

### Train the model

```python
#Train the model
clf.fit(X_train,y_train.ravel())
```

### make a prediction

```python
#make a prediction
y_pred=clf.predict(X_test)
```

### Evaluate the model on the training data and test data

```python
clf.score(X_test,y_test)
```

```python
clf.score(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
confusion = confusion_matrix(y_test, y_pred)
  
print ("Confusion Matrix : \n", confusion)
```

```python
print(classification_report(y_test,y_pred))
```

### Get the Accuracy score

```python
#Get the accuracy score
print ("Accuracy : ", accuracy_score(y_test, y_pred))
```

> A score of 0.82 means that the model has correctly predicted the outcome of 82% of the cases in the test set.

```python
clf.score(X_test,y_test)
```

> A score of 0.827 means that the model has correctly predicted the outcome of 82% of the cases in the test set.

### Improve the model by using different number of estimators

```python
import numpy as np
np.random.seed(40)
collect=[]
for i in range(10,100,10):
    print(f"Trying model with {i} estimators")
    model=RandomForestClassifier(n_estimators=i)
    model.fit(X_train,y_train.ravel())
    best='{:.2f}'.format(model.score(X_test,y_test) * 100)
    collect.append(best)
    print(f"Model accuracy on test set: {best} %")
    print(" ")
    
float_collect=np.asarray(collect,dtype=float)
print(f"The highest model accuracy is {np.max(float_collect)}")  
```

> After using different estimators ,90 produced a model with highest accuracy score
